```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('ip.jpg').convert('L')
pixels = img.load()

width, height = img.size

kernel = [[1/25 for j in range(5)] for i in range(5)]

output = Image.new('L', (width, height))
out_pixels = output.load()

for x in range(2, width-2):
    for y in range(2, height-2):
        total = 0

        for i in range(-2,3):
            for j in range(-2,3):
                nilai_pixel = pixels[x+i, y+j]
                nilai_kernel = kernel[i+2][j+2]
                total += nilai_pixel * nilai_kernel

        out_pixels[x,y] = int(total)

plt.imshow(output, cmap='gray')
plt.title("Hasil Konvolusi Kernel Rataan 5x5")
plt.axis('off')
plt.show()
```

Perulangan for x dan for y digunakan untuk menelusuri setiap piksel pada gambar, dimulai dari indeks 2 hingga width-2 dan height-2 agar kernel 5×5 tetap berada di dalam batas gambar. Di dalamnya, variabel total digunakan untuk menyimpan hasil sementara perhitungan konvolusi. Perulangan for i dan for j kemudian mengakses piksel di sekitar titik pusat (x, y) sesuai ukuran kernel. Setiap nilai piksel dikalikan dengan nilai kernel yang sesuai, lalu hasilnya dijumlahkan ke dalam total. Setelah semua elemen kernel diproses, nilai total disimpan sebagai nilai piksel baru pada gambar output (out_pixels[x,y]). Proses ini membuat setiap piksel menjadi rata-rata dari 25 piksel di sekitarnya sehingga menghasilkan efek penghalusan pada citra.
