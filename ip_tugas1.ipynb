```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/ip.jpg")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Gambar Asli")
plt.axis("off")
plt.show()
```

```python
def zoom_in(image, scale):
    height, width, channel = image.shape

    new_height = int(height * scale)
    new_width = int(width * scale)

    zoomed = np.zeros((new_height, new_width, channel), dtype=np.uint8)

    for i in range(new_height):
        for j in range(new_width):
            zoomed[i, j] = image[int(i/scale), int(j/scale)]

    return zoomed


zoomed_in = zoom_in(img_rgb, 2)

plt.figure(figsize=(8,6))
plt.imshow(zoomed_in)
plt.title("Zoom In 2x")
plt.axis("off")
plt.show()
```

```python
def zoom_out(image, scale):
    height, width, channel = image.shape

    new_height = int(height * scale)
    new_width = int(width * scale)

    zoomed = np.zeros((new_height, new_width, channel), dtype=np.uint8)

    for i in range(new_height):
        for j in range(new_width):

            x_start = int(i/scale)
            y_start = int(j/scale)

            x_end = min(x_start + int(1/scale), height)
            y_end = min(y_start + int(1/scale), width)

            block = image[x_start:x_end, y_start:y_end]

            zoomed[i, j] = np.mean(block, axis=(0,1))

    return zoomed


zoomed_out = zoom_out(img_rgb, 0.5)

plt.figure(figsize=(4,3))
plt.imshow(zoomed_out)
plt.title("Zoom Out 0.5x")
plt.axis("off")
plt.show()
```
